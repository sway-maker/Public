```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.edge.service import Service
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import cs
import time
# æŒ‡å®š Edge WebDriver çš„è·¯å¾‘
PATH = "C:\python\msedgedriver.exe"
service = Service(PATH)

# åˆå§‹åŒ– Edge WebDriver
driver = webdriver.Edge(service=service)
driver.get("https://rate.bot.com.tw/xrt")

# ç­‰å¾…ä¸»é è¡¨æ ¼åŠ è¼‰
WebDriverWait(driver, 10).until(
    EC.presence_of_element_located((By.XPATH, "/html/body/div[1]/main/div[4]/div/table"))
)

rows = driver.find_elements(By.XPATH, "//td[@class='currency phone-small-font']")
row_count = len(rows)
print(f"âœ… æ‰¾åˆ° {row_count} è¡Œæ•¸æ“š")

info_list = []

# æ‰¾åˆ°æ‰€æœ‰è²¨å¹£çš„ "æ­·å²æŸ¥è©¢" æŒ‰éˆ•
history_buttons = driver.find_elements(By.XPATH, '//*[@id="ie11andabove"]/div/table/tbody/tr/td[7]/a')
print(f"ğŸ” æ‰¾åˆ° {len(history_buttons)} å€‹å¹£åˆ¥")

for i in range(len(history_buttons)):
    try:
        # **é‡æ–°ç²å–æŒ‰éˆ•ï¼Œå› ç‚ºé é¢å¯èƒ½åˆ·æ–°**
        history_buttons = driver.find_elements(By.XPATH, '//*[@id="ie11andabove"]/div/table/tbody/tr/td[7]/a')
        
        # é»æ“Šç•¶å‰å¹£åˆ¥çš„æ­·å²æŸ¥è©¢æŒ‰éˆ•
        history_buttons[i].click()
        time.sleep(1)

        # åˆ‡æ›åˆ°æ–°é–‹å•Ÿçš„åˆ†é 
        WebDriverWait(driver, 10).until(lambda d: len(d.window_handles) > 1)
        driver.switch_to.window(driver.window_handles[-1])

        # é»æ“ŠæŸ¥è©¢æŒ‰éˆ•
        search_button = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.XPATH, '//*[@id="single"]'))
        )
        search_button.click()
        time.sleep(1)

        # ç­‰å¾…è¡¨æ ¼è¼‰å…¥
        WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.XPATH, '//*[@id="ie11andabove"]/div/table'))
        )

        # å–å¾—è²¨å¹£åç¨±
        currency_name = driver.find_element(By.XPATH, '//*[@id="ie11andabove"]/div/div[2]/div[2]/div/div[2]').text.strip()
        print(f"âœ… è²¨å¹£åç¨±: {currency_name}")

        # å–å¾—æ‰€æœ‰è¡¨æ ¼è¡Œæ•¸
        rows = driver.find_elements(By.XPATH, '//*[@id="ie11andabove"]/div/table/tbody/tr')
        print(f"âœ… æ‰¾åˆ° {len(rows)} ç­†æ•¸æ“š")

        # éæ­·æ¯ä¸€è¡Œï¼ŒæŠ“å–æ•¸æ“š
        for row in rows:
            datetime = row.find_element(By.XPATH, './td[1]').text.strip()
            buy_price = row.find_element(By.XPATH, './td[3]').text.strip()
            sell_price = row.find_element(By.XPATH, './td[4]').text.strip()

            print(f"âœ… æ›´æ–°æ™‚é–“: {datetime}, è²·å…¥åƒ¹: {buy_price}, è³£å‡ºåƒ¹: {sell_price}")

            # å­˜å…¥ list
            info_list.append({
                "è²¨å¹£åç¨±": currency_name,
                "æ›´æ–°æ™‚é–“": datetime,
                "è²·å…¥åƒ¹": buy_price,
                "è³£å‡ºåƒ¹": sell_price
            })

        # é—œé–‰ç•¶å‰åˆ†é ä¸¦åˆ‡å›ä¸»é 
        driver.close()
        driver.switch_to.window(driver.window_handles[0])
        print("âœ… è¿”å›ä¸»é ")

        time.sleep(2)  # ç­‰å¾…ä¸»é é‡æ–°è¼‰å…¥

    except Exception as e:
        print(f"âŒ ç™¼ç”ŸéŒ¯èª¤: {e}")
        driver.back()
        time.sleep(2)

# é—œé–‰ç€è¦½å™¨
driver.quit()
```
